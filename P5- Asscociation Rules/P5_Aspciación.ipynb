## Apriori

Calculamos los itemsets frecuentes

```
from mlxtend.frequent_patterns import apriori

# cambiar valores
frequent_itemsets_ap = apriori(prestamo, min_support=0.01, use_colnames=True)
print(frequent_itemsets_ap)
```

Obtenemos las reglas de asociaci칩n

```
from mlxtend.frequent_patterns import association_rules

rules_ap = association_rules(frequent_itemsets_ap, metric="confidence", min_threshold=0.75)
len(rules_ap)
```

Ordenamos seg칰n el lift

```
pd.set_option('display.max_rows', None)
rules_ap.sort_values(by='lift', ascending=False)
```

Ordenamos seg칰n la confianza

```
rules_ap.sort_values(by='confidence', ascending=False)
```

```
import seaborn as sns

fig, ax = plt.subplots(figsize=(7,7))
sns.scatterplot(x=rules_ap.support, 
                y=rules_ap.confidence, 
                size=rules_ap.lift, 
                sizes=(10, 200),
                hue=rules_ap.lift, 
                ax=ax).set_title("REGLAS DE ASOCIACI칍N")
```
